```python
# Group Publisher and sum number of positive ratings
publisher_df_positive = Df.groupby('publisher')[['positive_ratings']].sum()
top10_publishers=publisher_df_positive.sort_values(by='positive_ratings', ascending=False)
top10_publishers.head(15)
```

```python
# Group Publisher and sum number of positive ratings
publisher_df_negative = Df.groupby('publisher')[['negative_ratings']].sum()
bottom10_publishers=publisher_df_negative.sort_values(by='negative_ratings', ascending=False)
bottom10_publishers.head(10)
```

```python
# Group game genres and sum number of positive ratings
genre_df_positive = Df.groupby('genres')[['positive_ratings']].sum()
top10_genres=genre_df_positive.sort_values(by='positive_ratings', ascending=False)
top10_genres.head(10)
```

```python
# Group game genres and sum number of positive ratings
genre_df_negative = Df.groupby('genres')[['negative_ratings']].sum()
bottom10_genres=genre_df_negative.sort_values(by='negative_ratings', ascending=False)
bottom10_genres.head(10)
```

```python
#top 20 games by positive ratings 
selected_columns=Df[["name", "positive_ratings"]]
top20_games=selected_columns.copy().sort_values(by='positive_ratings', ascending=False)
top20_games.head(20)
```

```python
#top 20 games by positive ratings 
selected_worstcolumns=Df[["name", "negative_ratings"]]
bottom20_games=selected_worstcolumns.copy().sort_values(by='negative_ratings', ascending=False)
bottom20_games.head(20)
```

```python
#Attempt at splitting the genre column
genre_df = Df['genres'].str.split(';', expand=True).rename(columns = lambda x: "genre_"+str(x+1))
genre_df.head(20)
```

```python
genre_df['name'] = Df['name']
```

```python
col_name = 'name'
first_col = genre_df.pop(col_name)
genre_df.insert(0, col_name, first_col)
```

```python
#checking that the dataframe has been created based off game name
genre_df.columns.tolist()
```

```python
#merging genre dataset with the original dataset 
merged_df = pd.merge(Df, genre_df, on='name')
merged_df
```

```python
#checking list of columns available in the dataset after the merge
list(merged_df.columns) 
```

```python
#creating an average rating column and adding it to the dataset
merged_df['average_rating']=merged_df[["positive_ratings", "negative_ratings"]].mean(axis=1)
merged_df
```
